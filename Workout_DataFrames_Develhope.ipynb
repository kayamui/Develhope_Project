```python
y
```

```python
plt.plot(x, y)
```

```python
y = [0,4,2,4,10,15,6,7,11,12]
```

```python
plt.plot(x, y)
```

```python
plt.figure(figsize=(10,5))
plt.title("Plot".title())
plt.plot(x, y, color = "turquoise", label = "TR")
plt.plot(x, 2 * x+y, color="purple", label = "US")
plt.xlabel("Time")
plt.legend() # Grafik içinde US ve TR yazılarını gösterir
plt.ylabel("Revenue")
plt.show()
```

```python
plt.scatter(x,y, marker = "v") # Tüm bölgelerdeki verileri gösterir
plt.scatter(y,x, marker = "*")
plt.show()
```
