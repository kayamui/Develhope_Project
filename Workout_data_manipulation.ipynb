```python
# in the context of big data we wouldnt want to load all the rows into memory to then do a for loop

# we use the concept of apply
#we want to extract the year component from the data

def complex_function(order_date):
  
  try:
    result = str(order_date)[6:10]
  except:
     result = None
  return result

#after defining the function always test in a row

complex_function('07/07/2017')
```

```python
# now we want to apply the function above row by row in a pandas manner

df['Year of Order'] = df['Order Date'].apply(complex_function)
```

```python
df['Year of Order']
```

```python
df['Year of Order'] = df['Order Date'].apply(lambda x: int(str(x[6:10])))
```

```python
df['Year of Order']
```

```python
for sale, profit in df[['Sales', 'Profit']].head().iterrows():
  print(sale, profit)
```

```python
# more complex usecase whereby our 'complex function' receives multiple inputs

def more_complex_function(sales, profit):
  return sales - profit
df['new column'] = df.apply(lambda x : more_complex_function(x['Sales'], x['Profit']), axis = 1)
```

```python
df.head()
```

```python
def even_more_complex_function(sales, profit, multiplier):
  return multiplier * (sales - profit)
df['multiplied new column'] = df.apply(lambda x: even_more_complex_function(x['Sales'], x['Profit'], 3), axis = 1)
```

```python
#@title Rolling Average
# With this code I say take the average of column with the other 4 columns at upside (at total 5 columns),
# because of that first 4 columns has NaN value
```

```python
df['Moving Average'] = df['Sales'].rolling(5).sum()
df.rename(columns = {'Moving Average': 'Moving Sum of Sales'}, inplace=True)
```

```python
df.head(10)
```

```python
#min_periods means if I dont have 5 rows than take minimum 2 of them

df['Moving Average of Sales'] = df['Sales'].rolling(5, min_periods=2).mean()
```
